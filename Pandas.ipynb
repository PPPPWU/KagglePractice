<h2>What is a Series?</h2>
<p>A Pandas Series is like a column in a table.</p>
<p>It is a one-dimensional array holding data of any type.</p>
<p>pandas.Series( data, index, dtype, name, copy)</p>

```python
import pandas as pd
```

```python
a = [1, 7, 2]
index = ['x', 'y', 'z']
myvar = pd.Series(a, index=index)
print(myvar)
print(myvar['y'])
```

```python
sites = {1: "Google", 2: "Runoob", 3: "Wiki"}

myvar = pd.Series(sites, index = [1, 2, 3, 4])

print(myvar)
```

```python
myvar.values
```

<h2>What is a DataFrame?</h2>
<p>A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional 
array, or a table with rows and columns.</p>

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
```

```python
data = [['Google', 10], ['Runoob', 12], ['Wiki', 13]]

# 创建DataFrame
df = pd.DataFrame(data, columns=['Site', 'Age'])

# 使用astype方法设置每列的数据类型
df['Site'] = df['Site'].astype(str)
df['Age'] = df['Age'].astype(float)

print(df)
```

```python
company=["A","B","C"]
departs=['D','E']
df=pd.DataFrame({
    "company":[company[x] for x in np.random.randint(0,len(company),10)],
    "salary":np.random.randint(5,50,10),
    "age":np.random.randint(15,50,10),
    "Dept":[departs[x] for x in np.random.randint(0,len(departs),10)]
})
df
```

```python
d = {'A':'aa', 'B':'bb', 'C':'cc'}
df['company'].map(d)
```

```python
df.groupby(by='company').groups
```

```python
df.groupby(by = ['company', 'Dept'])['salary'].mean()
```

```python
temp = df.groupby(by='company', as_index=False)['salary'].mean()
new_temp  = pd.merge(df,temp, on='company')
new_temp
```

```python
dic = df.groupby(by='company')['salary'].mean().to_dict()
df['ave_sal'] = df['company'].map(dic)
dic, df
```

```python
new_temp  = pd.merge(df,temp, on='company')
new_temp
```

```python
def myfunc(p):
    sum = 0
    for i in p:
        sum += i
    return sum/len(p)
ap = df.groupby(by='company')['salary'].apply(myfunc)
tr = df.groupby(by='company')['salary'].transform(myfunc)
ap, tr
```
